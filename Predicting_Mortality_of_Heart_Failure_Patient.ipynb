# **Step 3: Data Preprocessing**

```python
x=data_df.drop(['DEATH_EVENT'], axis=1)
y=data_df['DEATH_EVENT']
```

```python
col_name=list(x.columns)
s_scaler= preprocessing.StandardScaler()
x_scaled=s_scaler.fit_transform(x)
x_scaled=pd.DataFrame(x_scaled, columns=col_name)
```

```python
x_scaled.describe().T
```

```python
plt.figure(figsize=(20,10))
sns.boxenplot(data=x_scaled)
```

```python
x_train,x_tests,y_train,y_test = train_test_split(x_scaled, y , test_size=0.3)
```

# **Step 4: Model Biulding**
### **1. svm**

```python
model1=svm.SVC()
```

```python
model1.fit(x_train,y_train)
```

```python
y_pred = model1.predict(x_tests)
```

```python
y_pred
```

```python
y_test
```

```python
print(classification_report(y_test, y_pred))
```

## **2. ANN**

```python
early_stopping = callbacks.EarlyStopping(min_delta= 0.001, patience= 20, restore_best_weights=True)
model=Sequential()
model.add(Dense(units=16, kernel_initializer = 'uniform' ,activation = 'relu', input_dim = 12))
model.add(Dense(units=8, kernel_initializer = 'uniform' , activation = 'relu'))
model.add(Dropout(0.25))
model.add(Dense(units=8, kernel_initializer = 'uniform' , activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units=1, kernel_initializer = 'uniform' , activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(x_train, y_train, batch_size=25, epochs=100, callbacks=[early_stopping], validation_split=0.2)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history_df = pd.DataFrame(history.history)
plt.plot(history_df[['loss']], label='training_loss')
plt.plot(history_df[['val_loss']], label='Validation_loss')
plt.legend()
plt.show()
```

```python
plt.plot(history_df.loc[:, ['accuracy']], label='training_accuracy')
plt.plot(history_df.loc[:, ['val_accuracy']], label='Validation_accuracy')
plt.legend()
plt.show()
```

```python
y_pred = model.predict(x_tests)
```

```python
y_pred = (y_pred > 0.5)
```

```python
print(classification_report(y_test, y_pred))
```
